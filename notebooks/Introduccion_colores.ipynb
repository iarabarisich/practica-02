# Cuerpo negro

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

from astropy.modeling.models import BlackBody
from astropy import units as u
from astropy.visualization import quantity_support

wav = np.arange(3000, 28000) * u.AA

def bb(wav,T):
    b = BlackBody ( temperature = T )
    flux = b(wav)
    return flux

plt.figure( figsize = (6, 4) )
for T in [3000*u.K, 8000*u.K]:
    plt.plot ( wav, bb(wav, T)/bb(wav, T).max(), label = 'T = {0}'.format(T) )

plt.xlabel(r'$\lambda$ [{0}]'.format(wav.unit) )
plt.ylabel(r'B($\lambda$, T) [{}]'.format(bb(wav,T).unit))

plt.legend()
plt.show()
```

**Pregunta:** ¿Cuál cuerpo negro emite mas radiación a longuitud de onda largas?

# Índice de color

```python
wav = np.arange(3000, 9000) * u.AA

def fill(a,T):
    bmax = bb(wav, T).max()
    range = np.arange(4250, 4900) * u.AA
    a.fill_between(range, 0, bb(range, T)/bmax, facecolor = 'blue', interpolate = True )
    range = np.arange(4900, 5600) * u.AA
    a.fill_between(range, 0, bb(range, T)/bmax, facecolor = 'green', interpolate = True )
    range = np.arange(6400, 7400) * u.AA
    a.fill_between(range, 0, bb(range, T)/bmax, facecolor = 'red', interpolate = True )
    
fig, (ax1, ax2, ax3) = plt.subplots ( 1, 3, figsize = (14, 4) )

T = 34000*u.K
fill(ax1,T)
ax1.plot ( wav, bb(wav, T)/bb(wav, T).max(), label = 'T = {0}'.format(T) )

T = 9790*u.K
fill(ax2,T)
ax2.plot ( wav, bb(wav, T)/bb(wav, T).max(), label = 'T = {0}'.format(T) )

T = 4410*u.K
fill(ax3,T)
ax3.plot ( wav, bb(wav, T)/bb(wav, T).max(), label = 'T = {0}'.format(T) )

for a in [ax1, ax2, ax3]:
    a.legend ( loc = 'upper center', bbox_to_anchor = (0.5, 1.2), ncol = 5 )
    a.set_xlabel ( r'$\lambda$ [{0}]'.format(wav.unit) )
    a.set_ylabel ( r'B($\lambda$, T) [{}]'.format(bb(wav,T).unit) )
    
fig.tight_layout() 
fig.show()
```

**Pregunta:** ¿Cuál cuerpo negro tiene (B-V) grande?

**A)** Estrellas caliente (T = 34.000K)
Las estrellas calientes emiten la mayor parte de su radiación en longitudes de onda cortas: B, U.

- Filtro B: mucha luz, por lo tanto, magnitud mas chica
- Filtro V: poca luz, por lo tanto, manigtud grande

En consecuencia, el índice $(B-V)$ < 0. Las estrella caliente tienen valores negativos del color $(B-V)$.

**B)** Temperaturas intermedias

**C)** Estrella fría (T = 4410K)

- Filtro B: poca luz, por lo tanto magnitud grande
- Filtro V: más luz, por lo tanto magnitud mas chica

En consecuencia, el índice $(B-V)$ > 0. Las estrella frías tienen valores positivos del color $(B-V)$.

    (B-V) color            temperature
    ----------------      ------------
    small   (< 0.5)          hot

     big    (> 2.0)          cool
    
El índice de color del Sol es aproximadamente $(B-V) = 0.6$.

**Pregunta:** ¿Cuáles son las principales diferencias entre los tipos espectrales?

```python
wav = np.arange(l1, l2) * u.AA

def plotS ( a, type ):
    df = cat[ (cat['lambdaAA'] >= l1) & (cat['lambdaAA'] <= l2) ]
    a.plot ( cat['lambdaAA'], cat[type]/df[type].max(), label = type )

fig, (ax1, ax2, ax3) = plt.subplots ( 1, 3, figsize = (14, 4) )

T = 30000*u.K
fill(ax3,T)
ax3.plot ( wav, bb(wav, T)/bb(wav, T).max(), label = 'T = {0}'.format(T) )
plotS (ax3, 'I(B0V)')

T = 9790*u.K
fill(ax2,T)
ax2.plot ( wav, bb(wav, T)/bb(wav, T).max(), label = 'T = {0}'.format(T) )
plotS (ax2, 'I(A3V)')

T = 4410*u.K
fill(ax1,T)
ax1.plot ( wav, bb(wav, T)/bb(wav, T).max(), label = 'T = {0}'.format(T) )
plotS (ax1, 'I(K5V)')

for a in ax1, ax2, ax3:
    a.set_xlim ( [ l1, l2 ] ) 
    a.set_ylim ( [-0.1, 1.1] ) 
    a.legend ( loc = 'upper center', bbox_to_anchor = (0.5, 1.2), ncol = 5 )
    a.set_xlabel ( r'$\lambda$ [{0}]'.format(wav.unit) )
    a.set_ylabel ( r'B($\lambda$, T) [{}]'.format(bb(wav,T).unit) )
    

fig.tight_layout() 
fig.show()
```

```python
from astroquery.vizier import Vizier

hipparcos = Vizier(catalog="I/239/hip_main", 
                   columns=['HIP', '_RAJ2000', '_DEJ2000', 'Plx','Vmag', 'B-V', 'SpType'],
                   row_limit = -1 ).query_constraints()[0]

hipparcos['M_V'] = hipparcos['Vmag'] + 5 * np.log10(hipparcos['Plx']/100.,
                                                    where = hipparcos['Plx'] > 0) 
```

```python
from PyAstronomy import pyasl

b = pyasl.BallesterosBV_T() # Tranformación (B-V) -> Teff

fig, ax1 = plt.subplots ( figsize = (9, 10) )

ax1.scatter ( hipparcos['B-V'], hipparcos['M_V'], s = 1, color = 'red', edgecolors = 'none' )

ax1.set_xlim ( -0.5,   2.5 )
ax1.set_ylim ( 15.0, -10.0 )
ax1.set_xlabel ( '(B-V)' )
ax1.set_ylabel ( 'V' )

ax2 = ax1.twiny()
ax2Xs = []
for X in ax1.get_xticks():
    Teff = b.bv2T(X)
    ax2Xs.append( "{0:4.2f}".format(Teff) )

ax2.set_xticklabels(ax2Xs)
ax2.set_xlabel ( 'T$_{eff}$ [K]' )

plt.show()
```
